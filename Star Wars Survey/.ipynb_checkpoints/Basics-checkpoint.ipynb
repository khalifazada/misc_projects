# Star Wars Survey

In this project we will be cleaning and exploring the data set from [FiveThirtyEight](https://github.com/fivethirtyeight/data/tree/master/star-wars-survey).
The question we will try to answer is if *The Empire Strikes Back* is the best of the sequels.

## Preparing the Dataset for Analysis

```python
import pandas as pd

star_wars = pd.read_csv('star_wars.csv', encoding='ISO-8859-1')
```

First, we'll need to remove the invalid rows.

```python
star_wars = star_wars[pd.notnull(star_wars['RespondentID'])]
```

## Cleaning And Mapping Yes/No Columns

We will use a dictionary to define a mapping from each value in a column to a new value:

```python
# mapping columns to convert to Boolean
values = {'Yes': True, 'No': False}

col_name = ['Have you seen any of the 6 films in the Star Wars franchise?',
           'Do you consider yourself to be a fan of the Star Wars film franchise?']

for name in col_name:
    star_wars[name] = star_wars[name].map(values)
```

## Cleaning And Mapping Checkbox Columns

The next six columns represent a single checkbox question. The respondent checked off a series of boxes in response to the question, *Which of the following Star Wars films have you seen? Please select all that apply.*

For each of these columns, if the value in a cell is the name of the movie, that means the respondent saw the movie. If the value is `NaN`, the respondent either didn't answer or didn't see the movie.

We'll convert each of these columns to a Boolean, then rename the column to `episode_n` where `n` will be the number of the episode.

```python
# list of column names
col_name = star_wars.columns[3:9]

# dictionary with values to map
map_values = {'nan': False}

# extract unique values for each column
unique_vals_lst = []
for name in col_name:
    tmp_lst = [str(item) for item in star_wars[name].unique()]
    unique_vals_lst.append(tmp_lst)

# add to the dict if not NaN then True
for vals_lst in unique_vals_lst:
    for item in vals_lst:
        if not (item == 'nan'):
            map_values[item] = True

# map values for each column
for name in col_name:
    star_wars[name] = star_wars[name].map(map_values)

# create dict with names 
rename_cols = {}
for i in range(0, len(col_name)):
    rename_cols[col_name[i]] = ('episode_' + str(i + 1))

# map column names to rename each
star_wars = star_wars.rename(columns=rename_cols)
```

## Cleaning The Ranking Columns

The next six columns ask the respondent to rank the *Star Wars* movies in order of least favorite to most favorite. 1 means the film was the most favorite, and 6 means it was the least favorite.

```python
# convert columns [9:15] to float
star_wars.iloc[:,9:15] = star_wars.iloc[:,9:15].astype(float)

# list of column names
col_name = star_wars.columns[9:15]

# create a map to rname columns [9:15]
rename_cols = {}
for i in range(0, len(col_name)):
    rename_cols[col_name[i]] = ('ranking_ep_' + str(i + 1))

star_wars = star_wars.rename(columns=rename_cols)
```

## Finding the Highest Ranked Episode

Now that we've cleaned up the ranking columns, we can find the highest-ranked movie more quickly.

```python
%matplotlib inline
import matplotlib.pyplot as plt

mean_rankings = star_wars.iloc[:,9:15].mean()

mean_rankings.plot.bar()

plt.show()
```

Seems like *Episode V* ranks as the best episode since lower score is better.

## Finding the Most Viewed Episode
Now we can figure out how many people have seen each movie.

```python
# sum of each column
total_viewed = star_wars[star_wars.columns[3:9]].sum()

total_viewed.plot.bar()
plt.show()
```

# Exploring The Data By Binary Segments

Let's examine how certain segments of the survey population responded. For example, we can split on the **`Gender`** column.

```python
male = star_wars[star_wars['Gender'] == 'Male']

male_most_viewed = male[male.columns[3:9]].sum()
male_most_viewed.plot.bar()
plt.title('Most Viewed Episode by Males')
plt.ylabel('View Count')
plt.xticks(rotation='horizontal')
plt.show()

male_highest_ranked = male[male.columns[9:15]].mean()
male_highest_ranked.plot.bar()
plt.title('Highest Score by Males \n(Lower is better)')
plt.ylabel('Score')
plt.show()
```

```python
female = star_wars[star_wars['Gender'] == 'Female']

female_most_viewed = female[female.columns[3:9]].sum()
female_most_viewed.plot.bar()
plt.title('Most Viewed Episode by Females')
plt.ylabel('View Count')
plt.xticks(rotation='horizontal')
plt.show()

female_highest_ranked = female[female.columns[9:15]].mean()
female_highest_ranked.plot.bar()
plt.title('Highest Score by Females \n(Lower is better)')
plt.ylabel('Score')
plt.show()
```

## Male/Female differences in favorite Star Wars movie and most seen movie

Interestingly, more males watches episodes 1-3, but males liked them far less than females did.
