# Earnings Based on College Majors
___

In this project I am working with a dataset on the job outcomes of students who graduated from college between 2010 and 2012.

The original data on these job outcomes was released by [American Community Survey](https://www.census.gov/programs-surveys/acs/), which conducts surveys and aggregates the data.

[FiveThirtyEight](https://fivethirtyeight.com/) cleaned the dataset and released it on their [GitHub](https://github.com/fivethirtyeight/data/tree/master/college-majors). 


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

# read the dataset
recent_grads = pd.read_csv('recent-grads.csv')

# examine the dataset
recent_grads.head(5)
```

```python
# dataset summary statistics
recent_grads.describe()
```

```python
# counting number of rows
raw_data_count = recent_grads.shape[0]

# drop any rows with missing data
recent_grads = recent_grads.dropna(axis=0)

# count number of rows again
cleaned_data_count = recent_grads.shape[0]

# compare values to find number of rows dropped
print('Raw Data Row Count: ', raw_data_count)
print('Cleaned Data Row Count: ', cleaned_data_count)
```

## Creating Plots

Below we will create plots of various columns to get a feel for the range and distribution of the data we are dealing with as well as determining some possible relationships (correlations) within the data.  

```python
ax1 = recent_grads.plot(x='Sample_size', y='Median', kind='scatter')
ax1.set_xlabel('Sample Size')
ax1.set_ylabel('Median')
ax1.set_title('Sample Size vs. Median')
```

```python
ax2 = recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter')
ax2.set_xlabel('Sample Size')
ax2.set_ylabel('Unemployment Rate')
ax2.set_title('Sample Size vs. Unemployment Rate')
```

```python
ax3 = recent_grads.plot(x='Full_time', y='Median', kind='scatter')
ax3.set_xlabel('Full Time')
ax3.set_ylabel('Median')
ax3.set_title('Full Time vs. Median')
```

```python
ax4 = recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter')
ax4.set_xlabel('Share of Women')
ax4.set_ylabel('Unemployment Rate')
ax4.set_title('Share of Women vs. Unemployment Rate')
```

```python
ax5 = recent_grads.plot(x='Men', y='Median', kind='scatter')
ax5.set_xlabel('Men')
ax5.set_ylabel('Median')
ax5.set_title('Men vs. Median')
```

```python
ax6 = recent_grads.plot(x='Women', y='Median', kind='scatter')
ax6.set_xlabel('Women')
ax6.set_ylabel('Median')
ax6.set_title('Women vs. Median')
```
